```python
#capturing the tables from the webpage to a dataframe 

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
postcode_temp=tables[0]
postcode=postcode_temp[postcode_temp.Borough != 'Not assigned']
postcode.reset_index(inplace = True, drop=True)
```

```python
#using pgeocode to find out coordinates using the postal codes and then saving thenm to a dataframe

geolocator = Nominatim(user_agent="test-user")
coord=[]
longlist=[]
latlist=[]
nomi = pgeocode.Nominatim("CA") #confirming the country, i.e. Canada

#using a for loop to run through the postcode dataframe and adding coordinates against each postal code

for i in postcode.index:        
    coord=nomi.query_postal_code(postcode.iloc[i]['Postal Code'])
    latlist.append(coord['latitude'])
    longlist.append(coord['longitude'])
postcode['Latitude']=latlist
postcode['Longitude']=longlist

#one location was not returned, at index 76. Here, the coordinates were found using Google and added manually at the index location.

postcode.at[76,'Latitude']= 43.6370
postcode.at[76,'Longitude']=-79.6158

postcode
```
